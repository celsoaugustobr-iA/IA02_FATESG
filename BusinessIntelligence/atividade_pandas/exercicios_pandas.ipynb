```python
import pandas as pd
import numpy as np

s_vazia = pd.Series(dtype='float64')
print("Series Vazia:")
print(s_vazia)
print("\nTipo de dado (dtype):", s_vazia.dtype)
```

```python
frutas = ['Maçã', 'Banana', 'Laranja', 'Uva', 'Manga']
s_frutas = pd.Series(frutas)
print(s_frutas)
```

```python
dados_numpy = np.array([10, 20, 30, 40])
s_indices = pd.Series(dados_numpy, index=['A', 'B', 'C', 'D'])
print(s_indices)
```

```python
dados_dict = {'jan': 100, 'fev': 200}
indices_personalizados = ['jan', 'fev', 'mar']

s_meses = pd.Series(dados_dict, index=indices_personalizados)
print(s_meses)
```
